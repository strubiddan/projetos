```python
import pandas as pd
import numpy as np
import yfinance as yf
import requests

df_planilha_acoes = pd.read_csv("planilha_acoes_2.csv", low_memory=False, encoding='latin1') # Carregando a planilha
display(df_planilha_acoes)
```

```python
######################## TRATANDO PLANILHA
df_planilha_acoes = df_planilha_acoes.rename(columns={'AÃÂÃÂO': 'AÃÃO'})
df_planilha_acoes = df_planilha_acoes.rename(columns={'PREÃÂO': 'PREÃO'})
df_planilha_acoes['AÃÃO'] = df_planilha_acoes['AÃÃO'].str.upper().str.strip().str.replace(' ','') # Padroniza o ticker para o padrÃ£o
df_planilha_acoes['PM'] = df_planilha_acoes['PM'].str.replace(',','.') 
df_planilha_acoes['PM'] = df_planilha_acoes['PM'].astype(float)        # Transforma o valor que estava sendo lido como string para float

############# TRATANDO OS INPUTS PARA USAR NA FUNÃÃO
ativos = df_planilha_acoes['AÃÃO'].tolist()
lista_ativos_renomeados =[]
for ativo in ativos:
    posicao = ativo.find('3')
    if posicao ==-1:
        posicao = ativo.find('4')
        if posicao ==-1:
            posicao = ativo.find('6')
            if posicao ==-1:
                    posicao = ativo.find('11')
    ativo = ativo+'.SA'
    lista_ativos_renomeados.append(ativo)

########################## GERANDO A PLANILHA
df_precos_fechamento = yf.download(lista_ativos_renomeados, period="1d")
df_precos_fechamento = df_precos_fechamento['Close']
display(df_precos_fechamento)
```

```python
#################### ATUALIZANDO OS PREÃOS
lista_valores = []
for ativo in lista_ativos_renomeados:
    valor = float(df_precos_fechamento[ativo].iloc[0])
    lista_valores.append(valor)
df_planilha_acoes['PREÃO'] = lista_valores
df_planilha_acoes['PREÃO'] = df_planilha_acoes['PREÃO'].round(2)
display(df_planilha_acoes)
df_planilha_acoes.to_csv('planilha_acoes2.csv')


```
